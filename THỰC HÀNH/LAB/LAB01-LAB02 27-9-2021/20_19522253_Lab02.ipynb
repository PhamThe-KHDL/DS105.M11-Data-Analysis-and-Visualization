## **Câu 8**: Dùng phương pháp thay thế (replace) để xử lý các giá trị bị khuyết cho các feature sau: `"normalized-losses"` (mean), `"stroke"` (choice), `"bore"` (mean), `"horsepower"` (choice), `"peak-rpm"` (mean). Và, `"num-of-doors"` (giá trị nhiều nhiều nhất).


```python
df_cau8 = df.copy()
df_cau8
```

```python
df_cau8.describe(include = 'all').transpose()
```

```python
# Thay thế các giá trị bị khuyết của các cột ['normalized-losses', 'bore', 'peak-rpm'] 
# bằng giá trị trùng bình (mean) của từng cột
df_cau8['normalized-losses'].fillna(df_cau8['normalized-losses'].mean().astype(np.int64), inplace=True)
df_cau8['bore'].fillna(df_cau8['bore'].mean(), inplace=True)
df_cau8['peak-rpm'].fillna(df_cau8['peak-rpm'].mean().astype(np.int64), inplace=True)
```

```python
# Thay thế các giá trị bị khuyết bằng giá trị xuất hiện nhiều nhất (mode)
df_cau8['num-of-doors'].fillna(df_cau8['num-of-doors'].mode(), inplace=True)
```

```python
# Thay thế các giá trị bị khuyết của các cột ['stroke', 'horsepower'] bằng giá trị lớn nhất (max) của từng cột
df_cau8['stroke'].fillna(df_cau8['stroke'].max(), inplace=True)
df_cau8['horsepower'].fillna(df_cau8['horsepower'].max(), inplace=True)
```

```python
df_cau8.to_csv('df_cau8.csv', index=False)
```

## **Câu 12**: Chuẩn hóa `"length"`, `"width"` và `"height"` trong phạm vi `[0,1]`.


```python
df_cau12 = df.copy()
df_cau12
```

```python
# Chuẩn hoá 'length' trong phạm vi [0,1]
df_cau12['length'] = (df_cau12['length'] - df_cau12['length'].min())/(df_cau12['length'].max() - df_cau12['length'].min())
```

```python
# Chuẩn hoá 'width' trong phạm vi [0,1]
df_cau12['width'] = (df_cau12['width'] - df_cau12['width'].min())/(df_cau12['width'].max() - df_cau12['width'].min())
```

```python
# Chuẩn hoá 'height' trong phạm vi [0,1]
df_cau12['height'] = (df_cau12['height'] - df_cau12['height'].min())/(df_cau12['height'].max() - df_cau12['height'].min())
```

## **Câu 13**: Tạo các bin là `['Low', 'Medium', 'High']` cho `“horsepower”` và `“price”`.

```python
df_cau13 = df.copy()
df_cau13
```

```python
# Xử lý các giá trị bị khuyết của 2 cột 'price' và 'horsepower' bằng các giá trị trung bình (mean) của mỗi cột
df_cau13['price'].fillna(df_cau13['price'].mean().astype('Int64'), inplace=True)
df_cau13['horsepower'].fillna(df_cau13['horsepower'].mean().astype('Int64'), inplace=True)
```

```python
df_cau13
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
bins_price = np.linspace(min(df_cau13['price']), max(df_cau13['price']), 4)
df_cau13['price_binned'] = pd.cut(df_cau13['price'], bins_price, labels=group_names, include_lowest=True)
```

```python
bins_horsepower = np.linspace(min(df_cau13['horsepower']), max(df_cau13['horsepower']), 4)
df_cau13['horsepower_binned'] = pd.cut(df_cau13['horsepower'], bins_horsepower, labels=group_names, include_lowest=True)
```

```python
from varname import argname2 

def handle_dfs(*dfs):
  names = argname2('*dfs')
  for name, df in zip(names, dfs):
    df.__dfname__ = name
     
  for df in dfs:
    return df.__dfname__
```

```python
def Save_File(file, file_name):
  file.to_csv(file_name + '.csv', index=False, header=True)
  file.to_excel(file_name + '.xlsx', sheet_name='Sheet_name_1', index=False)
  file.to_json(file_name + '.json', orient='index')
  # file.to_hdf(file_name + '.h5', key='df') # Lỗi chưa fix được
```
