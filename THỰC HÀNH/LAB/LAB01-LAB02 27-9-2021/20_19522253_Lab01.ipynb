Convert các kiểu dữ liệu về đúng định dạng

```python
df_cau6.replace(['?', '_', '.', ' ', 'Na', 'NULL', 'null', 'not'], np.NaN, inplace=True)
```

```python
df_cau6
```

```python
# Convert các cột ['normalized-losses', 'horsepower', 'peak-rpm', 'price'] 
# từ kiểu dữ liệu object sang kiểu dữ liệu Int64.
Convert_Int64 = ['normalized-losses', 'horsepower', 'peak-rpm', 'price']
for col in Convert_Int64:
  df_cau6[col] = pd.to_numeric(df_cau6[col], errors='coerce').astype('Int64')
```

```python
df_cau6
```

```python
# Convert các cột ['bore', 'stroke'] từ kiểu dữ liệu object sang kiểu dữ liệu float64.
Convert_float64 = ['bore', 'stroke']
for col in Convert_float64:
  df_cau6[col] = pd.to_numeric(df_cau6[col], errors='coerce')
```

```python
df_cau6
```

```python
df_cau6.info()
```

```python
df_cau6.to_csv('df_cau6.csv', index=False, header=True)
```
