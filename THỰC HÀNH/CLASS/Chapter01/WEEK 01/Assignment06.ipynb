## Các kiểu dữ liệu và phạm vi trong Python

````python
import sys
````

### **Numeric Types**: int, float and complex

Kiểu int: Kiểu số nguyên (không có chứa dấu chấm thập phân), có thể lưu các số nguyên âm và dương. 

````python
# Ví dụ
x = 123
x1 = -321

print("x = {}\nx1 = {}".format(x, x1))
print(type(x))
````

````python
print(sys.int_info)
````

Kiểu float: Kiểu số thực (có chứa dấu chấm thập phân)

````python
# Ví dụ
x2 = 3.1415

print("x2 = {}".format(x2))
print(type(x2))
````

````python
print(sys.float_info)
````

Kiểu complex: Kiểu số phức

````python
# Ví dụ
z = complex(2,3) # z = 2 + 3i

print("Phần thực = {} \nPhần ảo = {}".format(z.real, z.imag))
print(type(z))
````

### **Boolean**: Kiểu luận lý, để lưu True hoặc False

````python
# Ví dụ 
T = True 
F = False 

print("T = {} \nF = {}".format(T, F))
print(type(T))
````

### **Str**: Kiểu chuỗi, để trong nháy đôi hoặc nháy đơn 

````python
# Ví dụ 
S = 'Hello World' 

print("S = {}".format(S))
print(type(S))
````

### **List**: Được xác định bằng cặp dấu []. Có thể chứa nhiều loại dữ liệu khác nhau và có thể thay đổi.

````python
# Ví dụ 
L = [1, 2, 'String', 3.14, [4, 5], (7, 8)]

print("L = {}".format(L))
print(type(L))
````

### **Tuple**: Được xác định bằng cặp dấu (). Tương tự như List nhưng không thể sửa đổi.

````python
# Ví dụ 
Tuples = (1, 2, 'String', 3.14, [4, 5], (7, 8))

print("Tuples = {}".format(Tuples))
print(type(Tuples))
````

### **Set**: Được xác định bằng cặp dấu {}. chứa một tập hợp các đối tượng duy nhất không có thứ tự.

````python
# Ví dụ 
Set1 = set('Hello World')
Set2 = {6,1,3,2,4,1,2,6,4,2,3,1}

print("Set1 = {}".format(Set1))
print("Set2 = {}".format(Set2))

print(type(Set1))
print(type(Set2))
````

### **Dict**: Gồm các cặp key - value. Key là duy nhất.

````python
# Ví dụ 
a = dict(one=1, two=2, three=3)
b = {'one': 1, 'two': 2, 'three': 3}
c = dict(zip(['one', 'two', 'three'], [1, 2, 3]))
d = dict([('two', 2), ('one', 1), ('three', 3)])
e = dict({'three': 3, 'one': 1, 'two': 2})

print(a == b == c == d == e)
print("a = {}".format(a))
print(type(a))
````

## Các kiểu dữ liệu và phạm vi trong Numpy

NumPy hỗ trợ nhiều kiểu dữ liệu hơn nhiều so với Python. Bảng sau đây cho thấy các kiểu dữ liệu vô hướng khác nhau được xác định trong NumPy.
```python
bool_ : Boolean (Đúng hoặc Sai) được lưu trữ dưới dạng byte
int_ : Kiểu số nguyên mặc định (giống C long; thường là int64 hoặc int32)
intc : Giống hệt với int C (thường là int32 hoặc int64)
intp : Số nguyên được sử dụng để lập chỉ số (giống như C ssize_t; thông thường là int32 hoặc int64)
int8 : Byte (-128 đến 127)
int16 : Số nguyên (-32768 đến 32767 )
int32 : Số nguyên (-2147483648 đến 2147483647 )
int64 : Số nguyên (-9223372036854775808 đến 9223372036854775807 )
uint8 : Số nguyên không dấu (0 đến 255)
uint16 : Số nguyên không dấu (0 đến 65535)
uint32 : Số nguyên không dấu (0 đến 4294967295)
uint64 : Số nguyên không dấu (0 đến 18446744073709551615)
float_ : Viết tắt cho float64
float16 : Nửa chính xác float: bit dấu, số mũ 5 bit, phần định trị 10 bit
float32 : số thực chính xác đơn: bit dấu, số mũ 8 bit, phần định trị 23 bit
float64 : số thực chính xác kép: bit dấu, số mũ 11 bit, phần định trị 52 bit
complex_ : Viết tắt cho complex128
complex64: Nó được sử dụng để biểu diễn số phức trong đó phần thực và phần ảo chia sẻ mỗi phần 32 bit.
complex128: Nó được sử dụng để biểu diễn số phức trong đó phần thực và phần ảo chia sẻ mỗi phần 64 bit.
```
