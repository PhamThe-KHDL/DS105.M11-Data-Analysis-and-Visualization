# Định nghĩa các hàm

## Định nghĩa hàm chuyển chuỗi nhập từ bàn phím thành 1 list

```python
def Convert(Ticket):
  return list(map(int, Ticket.split('-')))
```

## Định nghĩa hàm kiểm tra các số của 1 vé có thoả mãn điều kiện khác nhau hay không?

```python
def Check_Ticket(Ticket):
  A = Convert(Ticket)
  for i in A:
    if i < 1 or i > 45:
      return False
  if len(A) != 6:
    return False
  B = set(A)
  if len(A) == len(B):
    return True
  return False
```

## Định nghĩa hàm random số trúng thưởng

```python
def Random_Number():
  results = []
  i = 0
  while(i<6):
    result = random.randrange(1,45)
    if result not in results:
      results.append(result)
      i += 1
  return results
```

## Định nghĩa hàm nhập mã số vé số

```python
def Input_Tickets(n):
  Tickets = []
  i = 0
  while True:
    try:
      while(i < n):
        print("Nhập mã số vé số thứ", i+1)
        Ticket = input("Nhập mã số vé số của bạn ví dụ: 10-32-14-25-21-15 ")
        if Check_Ticket(Ticket):
          Tickets.append(Convert(Ticket))
          i +=1
        else:
          print("Không thoả, nhập lại!")
    except:
      print("Nhập lại!")
    if i == n:
      break
  return Tickets
```

## Định nghĩa hàm kiểm tra số tiền còn lại sau mỗi lượt chơi

```python
def Check_Money(Money, PurchaseCost, bonus):
  Money = Money - PurchaseCost + bonus
  return Money 
```

## Định nghĩa hàm so sánh giữa vé số trúng thưởng và vé số của người chơi

```python
def Compare(Tickets, result):
  Counts = []
  for i in range(len(Tickets)):
    count = len(set(Tickets[i]) & set(result))
    Counts.append(count)
  return Counts
```

## Định nghĩa hàm tiền thưởng cho người chơi trúng thưởng

```python
def Bonus(compares):
  BONUS = (30000, 300000, 10000000, 10000000000)
  bonus = []
  for i in compares:
    if i < 3:
      bonus.append(0)
    elif i == 3:
      bonus.append(BONUS[0])
    elif i == 4:
      bonus.append(BONUS[1])
    elif i == 5:
      bonus.append(BONUS[2])
    elif i == 6:
      bonus.append(BONUS[3])
  return bonus
```

## Định nghĩa hàm tóm tắt quá trình chơi vé số Vietlot

```python
def Summary(StartMoney, SumCost, SumBonus, count, MaxBonus, Money):
  # Số tiền ban đầu 
  print("Số tiền ban đầu: {} đồng".format(StartMoney))
  # Tổng số tiền đã bỏ ra để mua vé số
  print("Tổng số tiền đã bỏ ra để mua vé số: {} đồng".format(SumCost))
  # Tổng số tiền đã trúng vé số    
  print("Tổng số tiền đã trúng vé số: {} đồng".format(SumBonus))
  # Số lượt chơi
  print("Số lượt chơi: {} lượt".format(count))
  # Giải cao nhất đã trúng
  print("Giải cao nhất đã trúng: {} đồng".format(MaxBonus))
  # Số tiền hiện tại
  print("Số tiền hiện tại: {} đồng".format(Money))
```

## Định nghĩa hàm nhập số lượng người chơi

```python
def Number_Players():
  while True:
    try:
      Players = int(input("Nhập số lượng người chơi (>0): "))
      if Players > 0:
        break
    except:
      print("Không hợp lệ, nhập lại!")
    print("Nhập lại!")
  return Players
```

## Định nghĩa hàm kiểm tra số lượng vé số cần mua

```python
def Number_Vietlot(Money):
  while True:
      Cost = 10000 # Giá của một tờ vé số
      while True:
        try:
          n = int(input("Nhập số lượng vé số cần mua:"))
          if n > 0:
            break
        except:
          print("Không hợp lê!")
      PurchaseCost = n*Cost # Số tiền bỏ ra để mua vé số
      # Kiểm tra xem có đủ tiền để mua vé số hay không
      CheckMoney = Check_Money(Money, PurchaseCost, 0)
      if CheckMoney < 0:
        print("Không đủ tiền!")
        print("Số tiền còn lại: {}".format(Money))
      else:
        break
  return n, PurchaseCost
```

## Định nghĩa hàm chơi vé số Vietlot

```python
def Play_Game():
  Money = [100000] # Danh sách số tiền ban đầu của mỗi người chơi và được cập nhật trong quá trình chơi
  SumCost = [0] # Danh sách tổng số tiền đã bỏ ra để mua vé của mỗi người chơi
  SumBonus = [0] # Danh sách tổng số tiền đã trúng vé số của mỗi người chơi
  Count = [0] # Danh sách số lượt chơi của người mỗi chơi
  MaxBonus = [0] # Danh sách giải cao nhất đã trúng của mỗi người chơi
  StartMoney = 100000 # Danh sách số tiền ban đầu của mỗi người chơi
  Tickets = [0] # Danh sách vé số của mỗi người chơi
  PurchaseCost = [0] # Danh sách số tiền mua vé số của mỗi người chơi trong một lươt chơi
  players = 0 # Số lượng người chơi
  flag = 1

  while True:
      print("-"*10 + "TRÒ CHƠI VÉ SỐ VIETLOT" + "-"*10)
      play = int(input("Nhập 1: Bắt đầu\nNhập 0: Kết thúc\n"))
      while play == 1:
        print("Trò chơi bắt đầu!")
        if flag:
          players = Number_Players()
          print("Số lượng người chơi:", players)
          Money = Money * players
          print("Danh sách tiền ban đầu của mỗi người chơi:", Money)
          Count = Count * players
          print("Danh sách số lượt chơi của mỗi người chơi:", Count)
          Tickets = Tickets * players
          print("Danh sách vé của mỗi người chơi:", Tickets)
          SumCost = SumCost * players
          print("Danh sách tổng số tiền đã bỏ ra để mua vé số của mỗi người chơi:", SumCost)
          SumBonus = SumBonus * players
          print("Danh sách tổng số tiền trúng thưởng vé số của mỗi người chơi:",SumBonus)
          MaxBonus = MaxBonus * players
          print("Danh sách số tiền trúng thưởng cao nhất của mỗi người chơi:",MaxBonus)
          PurchaseCost = PurchaseCost * players
          print("Danh sách số tiền mua vé số mỗi người chơi:",PurchaseCost)
          flag = 0

        for i in range(players):
          print("Người chơi thứ", i+1)
          # Kiểm tra xem người chơi còn tiền để tiếp tục chơi hay không? Nếu không còn tiền sẽ dừng lượt chơi 
          if Money[i] == 0:
            print("Bạn đã hết tiền, tạm biệt!")
            continue
          n, PurchaseCost[i] = Number_Vietlot(Money[i])
          print("Số lượng vé số đã mua:", n)
          print("Số tiền bỏ ra để mua vé số:", PurchaseCost[i])
          Count[i] += 1
          print("Số lượt chơi của người chơi thứ {}: {}".format(i+1, Count[i]))
          SumCost[i] += PurchaseCost[i]
          print("Tổng số tiền đã bỏ ra để mua vé số của người chơi thứ {}: {}".format(i+1, SumCost[i]))
          Tickets[i] = Input_Tickets(n)
          print("Danh sách vé số của người chơi thứ {}: {}".format(i+1, Tickets[i]))
        
        Result = Random_Number() # Kết quả sổ số
        print("Mã số trúng thưởng:", Result)
        for i in range(players):
          if Money[i] == 0:
            continue
          print("Danh sách vé số của người chơi thứ {} là: {}".format(i+1, Tickets[i]))
          Compares = Compare(Tickets[i], Result)
          print("Số kết quả giống nhau của người chơi thứ {} là: {}".format(i+1, Compares))
          bonus = sum(Bonus(Compares))
          SumBonus[i] += bonus
          if MaxBonus[i] < max(Bonus(Compares)):
            MaxBonus[i] = max(Bonus(Compares))
        
          print("Tiền thưởng trúng vé số của người thứ {} là {} đồng".format(i+1, bonus))
          # Cập nhật lại số tiền đang có sau khi công bố giải thưởng
          Money[i] = Check_Money(Money[i], PurchaseCost[i], bonus)
          print("Số tiền còn lại của người thứ {} là: {} đồng".format(i+1, Money[i]))

        play = int(input("Nhập 1: Tiếp tục\nNhập 0: Kết thúc\n"))

      if play == 0:
        print("Trò chơi kết thúc!")
        break
      else:
        print("Nhập lại!")
      
  for i in range(players):
    print("-"*10 + "TÓM TẮT QUÁ TRÌNH CHƠI VÉ SỐ VIETLOT CỦA NGƯỜI CHƠI THỨ {}".format(i+1) + "-"*10)
    Summary(StartMoney, SumCost[i], SumBonus[i], Count[i], MaxBonus[i], Money[i])
```
